```python
from zimg import ZImg

# data_file = r'D:\EEG Data\niiData\20230426\002_slice_0020_fused_tp_0_ch_0_cropped_v2n4corr.nii.gz'
data_file = r"D:\EEG Data\niiData\20230530\full16_100um_2009b_sym.nii.gz"
zimg_image = ZImg(data_file)
zimg_data = zimg_image.data[0]
```

```python
import matplotlib.pyplot as plt
import math

single_image = zimg_data[0, 0, :, :]
vmin = max(0, math.floor(single_image.min()))
vmax = min(255, math.ceil(single_image.max()))
print((vmin, vmax))
plt.imshow(single_image, cmap="gray", vmin=vmin, vmax=vmax)
```

```python
import numpy as np

integer_image = single_image.astype(np.uint8)
plt.imshow(integer_image, cmap="gray")
```

```python
def normalize(array: np.ndarray) -> np.ndarray:
    max_data = array.max()
    min_data = array.min()
    return (array - min_data) / (max_data - min_data)


normalized_image = normalize(single_image)
plt.imshow(normalized_image, cmap="gray", vmin=0.0, vmax=1.0)
```
